1. Definition of Summary Statistics: <br>
  a) Count: It tells you how many non-null (non-missing) values are present in each column. It counts the number of values available for statistical analysis.
  
  b) Mean: Mean is referred to as the arithmetic average and is the average value of the data points in the column. Its Calculation: mean= (sum of all the values) / count of non-missing values. df ['x'].mean().
  
  c) Standard Deviation (std): Provides a measure of the spread of data points around the mean. Low std is interpreted as the values being closer to the mean, while a higher one indicates that the values are more spread-out. df ['x'].std()
  
  d) Minimum (min): The smallest value represented in the column for a datset. df ['x'}.min ().
  
  e) 25% (1st quartile):The value below which 25% of the data fall. df['x'].quantile(0.25)
  
  f) 50% (median or 2nd quartile): The middle value in the data set--half of the data falls below this point and half above. df['x'].quantile(0.5)
  
  g) 75% (3rd quartile): The value below which 75% of the data fall. df['x'].quantile(0.75)
  
  h) Maximum (max): The largest value represented in the column for a data set. df ['x'].max().

8.1 The df.groupby('col1') function groups the DataFrame df using the values in ('col1'). The ["col2"].describe() function generates summary statistics for the ('col2') such as std, mean, count, 50%, 25%, etc.

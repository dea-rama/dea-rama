```python
#3
import numpy as np #Imprts NumPy library which is used for random number generation

all_door_options = [1, 2, 3] #tuple #Three door options (1,2,or 3)
my_initial_choice = 1  # Player always starts by picking door 1
wins = 0  #Counter that keeps track of how many times the player wins
reps = 100000 # Number of times that the Monty Hall simulation is run

for _ in range(reps): #The loop will run reps or 100000 times
    # Randomly assign the car to one of the doors
    secret_winning_door = np.random.choice(all_door_options)
    
    # Monty reveals a losing door (not the player's choice and not the winning door)
    remaining_doors = [door for door in all_door_options if door != my_initial_choice and door != secret_winning_door]
    monty_reveal = np.random.choice(remaining_doors)
    
    # Player switches to the remaining door
    switch_choice = [door for door in all_door_options if door != my_initial_choice and door != monty_reveal][0]
    
    # Checks if the switched choice is the winning door
    if switch_choice == secret_winning_door:
        wins += 1

#Probability of winning when switching
win_rate = wins / reps
print(f"Win rate when switching: {win_rate:.2f}")
```

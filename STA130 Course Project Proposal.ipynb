### Analysis Proposal I: Relationship between Social Activities with Family Members and Mental Well-Being

**1. Research Question**: How does the frequency of visiting family members relate to Canadians' mental health?

**2. Variables:** 


**a) CONNECTION_activities_visited_family_p3m** (Independent Variable)

This variable categorizes the frequency of family visits in the past 3 months on a scale of daily to not in the past three months. The responses from each individual provide a clear understanding of the frequency of their social engagement, which can provide valuable information when analyzing the relationship with their mental health ratings.

**b) WELLNESS_self_rated_mental_health** (Dependent Variable)

This variable measures individuals' perception of their own mental health on a scale from poor to excellent. Because the scale is very detailed, it will provide detailed results on how social interactions impact mental health. This variable also highlights how mental health varies among different groups of people, which is important for my analysis of the overall population. 


**Visualization Method: Bar Plot**

Since the data I am using is ordinal (categorical), the best visualization method would be a bar plot, which will effectively display the average self-rated mental health scores for each frequency category of family member interactions and provide a clear comparison of mental health score variations across different family interaction frequencies.



 

**3. Analysis:**

**Method: Simple Linear Regression** to analyze the relationship between the frequency of visiting family members and self-rated mental health. This will help determine the extent to which family visits account for the variation in mental health.

**Process**: In my Simple Linear Regression model, I will set the CONNECTION_activities_visited_family_p3m to be the independent variable and WELLNESS_self_rated_mental_health to be the dependent variable, which will demonstrate how self-rated mental health depends (if it does) on the frequency of family visits. Then, I will encode the ordinal data of both variables by assigning numbers to the mental health ratings and family visit frequencies. Next, I will build the simple linear regression model and analyze the regression coefficients like the slope to determine if an increase in visit frequencies improves mental health, and use the y-intercept as the baseline mental health level for the lowest frequency of social visits.

**Assumptions**: The relationship between the frequency of family visits (independent variable) and self-rated mental health scores (dependent variable) is linear; each observation is linearly independent; the errors are normally distributed, homoscedastic, and statistically independent; and the sample is a representation of the Canadian population.

**4.Null Hypothesis:** There is no significant relationship (slope = 0) between family visit frequency and individuals' self-rated mental health scores.

**Alternative Hypothesis:** There is a positive relationship between family visit frequency and self-rated mental health scores.

I believe that my linear regression model will reject the null hypothesis in favour of the alternative hypothesis, showing a positive linear slope. This would indicate that increased frequency of family visits increases the self-rated mental health scores, directly addressing the research question—"How does the frequency of visiting family members relate to the mental health of Canadians?." If confirmed by statistical evidence, the results will highlight the importance of maintaining familial connections for mental health and address one of the course project's objectives.
